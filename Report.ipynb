<span style="font-size:24pt">Построить верхнюю и нижнюю суммы Дарбу для равномерного разбиения на n частей.</span>

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 1.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Определение размера подинтервала</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Интервал $[0, 1]$ разбивается на $n$ равных частей, так что каждый подинтервал имеет размер:</span>
<div style="text-align: center; font-size:16pt;"><br>
    $\Delta x = \frac{b \, - \, a}{n} = \frac{1 \, - \, 0}{n} = \frac{1}{n}$
</div>

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 2.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Точки разбиения и значения функции</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Точки разбиения задются так:</span>
<div style="text-align: center; font-size:16pt;"><br>
    $x_{i} = 1 + i \cdot \Delta x,\, где \,\,i = 0, 1, ..., n$

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 3.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Нижняя и верхняя суммы Дарбу</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Для нижней суммы Дарбу $s_{t}$ на каждом подинтервале берется минимальное значение функции, что для $f(x) = e^{x}$ на нашем интервале будет в начале каждого подинтервала. Таким образом</span>
<div style="text-align: center; font-size:16pt;"><br>
    $m_{i} = \inf\limits_{x \in \Delta i} f(x) = f\left(\min\limits_{x \in \Delta i} x\right) = f(x_{i-1})$ так как $f(x) \uparrow$ на $x \in [0, 1]$
    
<span style="font-size:12pt; margin-bottom: 20px;">Для верхней суммы Дарбу $S_{t}$ на каждом подинтервале берется максимальное значение функции, что для $f(x) = e^{x}$ на нашем интервале будет в конце каждого подинтервала. Таким образом</span>
<div style="text-align: center; font-size:16pt;"><br>
    $M_{i} = \sup\limits_{x \in \Delta i} f(x) = f\left(\max\limits_{x \in \Delta i} x\right) = f(x_{i})$ так как $f(x) \uparrow$ на $x \in [0, 1]$

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 4.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Формулы для сумм</span>
<div style="margin-bottom: 12px;"></div>

<div style="text-align: center; font-size:16pt; margin-bottom: 30px;"><br>
    $s_{t}(f) = \sum_{i=1}^{n} f(x_{i-1}) \cdot \Delta x_{i} = \sum_{i=1}^{n} f\left(\frac{i-1}{n}\right) \cdot \frac{1}{n} = 
    \frac{1}{n} \sum_{i=0}^{n-1} e^\frac{i}{n} = \frac{1 \cdot (e-1)}{n \cdot (e^\frac{1}{n} - 1)}$
    
<div style="text-align: center; font-size:16pt; margin-bottom: 30px;"><br>
    $S_{t}(f) = \sum_{i=1}^{n} f(x_{i}) \cdot \Delta x_{i} = \sum_{i=1}^{n} f\left(\frac{i}{n}\right) \cdot \frac{1}{n} = 
    \frac{1}{n} \sum_{i=1}^{n} e^\frac{i}{n} = \frac{e^\frac{1}{n} \cdot (e-1)}{n \cdot (e^\frac{1}{n} - 1)}$

<span style="font-size:24pt">Доказать, что функция интегрируема по Риману.</span>

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 1.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Непрерывность функции</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Функция $f(x) = e^{x}$ ограничена на $[0, 1]$, значит суммы Дарбу конечны</span>


<span style="font-size:18pt; margin-bottom: 10px;">Шаг 2.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Критерий Дарбу</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Чтобы доказать интегрируемость по Риману, нужно показать, что разность между верхней и нижней суммами Дарбу стремится к нулю при увеличении числа разбиений $n$, то есть при $n \to \infty$</span>
<div style="text-align: center; font-size:16pt;"><br>
    $\lim\limits_{\frac{1}{n} \to 0} (S_{t}(f) - s_{t}(f)) = \lim\limits_{n \to \infty} \frac{(e^\frac{1}{n}-1)(e-1)}{n \cdot (e^\frac{1}{n}-1)} = 0$
    
<span style="font-size:12pt; margin-bottom: 20px;">Увидим, что разность между верхней и нижней суммами Дарбу может быть сделана сколь угодно малой, и следовательно, функция $f(x) = e^{x}$ Интегрируема по Риману на отрезке $[0, 1]$</span>


<span style="font-size:24pt">Найти пределы сумм Дарбу, найти значения интеграла.</span>

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 1.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Пределы сумм Дарбу</span>
<div style="margin-bottom: 12px;"></div>

<div style="text-align: center; font-size:16pt;"><br>
    $\lim\limits_{n \to \infty} s_{t}(f) = \lim\limits_{n \to \infty} \frac{e-1}{n \cdot (e^\frac{1}{n} - 1)} = \lim\limits_{n \to \infty} \frac{e-1}{n \cdot \frac{1}{n}} = e-1$
<div style="text-align: center; font-size:16pt;"><br>
    $\lim\limits_{n \to \infty} S_{t}(f) = \lim\limits_{n \to \infty} \frac{e^\frac{1}{n} \cdot (e-1)}{n \cdot (e^\frac{1}{n} - 1)} = \lim\limits_{n \to \infty} \frac{e^\frac{1}{n} \cdot (e-1)}{n \cdot \frac{1}{n}} = e-1$
    

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 2.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Значения интеграла</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Так как пределы сумм Дарбу совпадают - мы можем утверждать что значение определенного интеграла равно</span>
<div style="text-align: center; font-size:16pt;"><br>
    $\lim\limits_{n \to \infty} s_{t}(f) = \lim\limits_{n \to \infty} S_{t}(f) = \int_0^1 e^x dx = e-1$
   

<span style="font-size:24pt">Сравнить со значением, полученным по формуле Ньютона-Лейбница.</span>

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 1.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Первообразные</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Формула Ньютона-Лейбница гласит:</span>
<div style="text-align: center; font-size:16pt;"><br>
    $\int_a^b f(x)dx = F(b) - F(a)$
    
<span style="font-size:12pt; margin-bottom: 20px;">Найдем первообразную для $f(x) = e^x$. Первообразная для функции $e^x$ есть $e^x$, получется:</span>
<div style="text-align: center; font-size:16pt;"><br>
    $F(x) = \int e^x dx = e^x + C$
   

<span style="font-size:18pt; margin-bottom: 10px;">Шаг 2.</span>
<span style="font-size:18pt; margin-bottom: 10px;">Применим формулу Ньютона-Лейбница</span>
<div style="margin-bottom: 12px;"></div>
<span style="font-size:12pt; margin-bottom: 20px;">Подставим $F(x) = e^x$ и пределы интегрирования 0 и 1:</span>
<div style="text-align: center; font-size:16pt;"><br>
    $\int_0^1 f(x)dx = e^1 - e^0 = e - 1$

<span style="font-size:18pt; margin-bottom: 10px;">Вывод</span>
